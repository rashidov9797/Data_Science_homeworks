```python

# cut va qcut yordamida guruhlash

df3= pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv",usecols=['country','pop2021','area'],index_col='country')
df3.head()
```

```python
df3.loc['Uzbekistan']
```

```python
df3.loc['India']
```

```python
mils = [0,1000,10000,30000,50000,100000,300000,500000]
```

```python
populations = df3.pop2021
```

```python
populations
```

```python
guruh = pd.cut(populations,mils)
guruh
# pd.cut -- populationdagi aholini milsdagi qiymatlarga qarab ajrat
# mils = [0,1000,10000,30000,50000,100000,300000,500000]
```

```python
guruh.value_counts()
```

```python
df3['guruh'] = guruh
# Guruh degan yangi ustun yaratib guruh degan jadvalni qoshdik
df3.head()
```

```python
# Avtomobillarni narxlarini guruhlash
narx = [0,1000,20000,50000]
```

```python
range = pd.cut(df.price,narx)
range
# price degan ustundagi narxlarni narx jadvali yordamida bn ajratib ber
```

```python
range.value_counts()
```

```python
names = ['arzon','qimmat',"o'rta"]
range = pd.cut(df.price,narx,labels=names)
# labellariga namesdagi nomlarni joyla
range.value_counts()
```

```python
df['guruh'] = range
df.tail()
```

```python

range = pd.cut(df.price,4)
# guruhlarni avtomatik bo'lish
```
